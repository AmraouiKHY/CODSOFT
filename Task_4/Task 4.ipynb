```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,accuracy_score,r2_score
%matplotlib inline 
```

```python
X = sales_data.drop(columns='Sales')
y= sales_data['Sales']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"Mean Absooolute Error: {mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test,y_pred)}")
print(f"R squared score: {r2_score(y_test,y_pred)}")
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```python
y_pred_reg = regressor.predict(X_test)
```

```python
print(f"Mean Absooolute Error: {mean_absolute_error(y_test,y_pred_reg)}")
print(f"Mean Squared Error: {mean_squared_error(y_test,y_pred_reg)}")
print(f"R squared score: {r2_score(y_test,y_pred_reg)}")
```
